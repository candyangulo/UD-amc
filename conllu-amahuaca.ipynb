```python
import os
import pandas as pd
from conllu.parser import ParseException

def preprocess_data(data):
    # Replace "nan" with a placeholder, e.g., "NAN_PLACEHOLDER"
    data = data.replace("nan", "NAN_PLACEHOLDER")
    return data

def postprocess_data(data):
    # Replace the placeholder back to "nan"
    data = data.replace("NAN_PLACEHOLDER", "nan")
    return data

file_path = r'C:\Users\ASUS\UD\treebank-amahuaca.conllu'
amahuaca = []

try:
    with open(file_path, "r", encoding="utf-8") as data_amc:
        data = data_amc.read()
        
        # Preprocess the data to replace "nan" temporarily
        preprocessed_data = preprocess_data(data)
        
        sentences = preprocessed_data.strip().split('\n\n')  # Split data into sentences

        for sentence_str in sentences:
            amahuaca.append(sentence_str)

        print(f"Processed file: {file_path}")
except Exception as e:
    print(f"Error processing file {file_path}: {e}")

# Now you can work with the 'amahuaca' data, and "nan" will be treated as a valid word
with open(r"G:\Other computers\Mi portátil\D\Proyectos\UD\Oraciones formato conllu\Amahuaca (amc)\treebank-amahuaca.conllu", "w", encoding='utf-8') as f:
    for sentence_data in amahuaca:
        # Postprocess the data to replace the placeholder back to "nan"
        sentence_data = postprocess_data(sentence_data)
        f.write(sentence_data)
        f.write('\n\n')
```

```python
# Initialize counters for dependency relations
relation_counts = {}
total_relations = 0

for sentence_data in amahuaca:
    # Postprocess the data to replace the placeholder back to "nan"
    sentence_data = sentence_data.replace("NAN_PLACEHOLDER", "nan")
    
    # Parse the sentence data into a Conllu structure
    sentence_data_conllu = parse(sentence_data)
    
    for sentence in sentence_data_conllu:
        for word in sentence:
            if 'deprel' in word:
                dep_rel = word['deprel']
                total_relations += 1
                if dep_rel in relation_counts:
                    relation_counts[dep_rel] += 1
                else:
                    relation_counts[dep_rel] = 1

# Calculate the percentage of each dependency relation
relation_percentages = {rel: (count / total_relations) * 100 for rel, count in relation_counts.items()}

# Print the results
print(f"Total number of dependency relations: {total_relations}")
print("Dependency relation counts:")
for rel, count in relation_counts.items():
    print(f"{rel}: {count} ({relation_percentages[rel]:.2f}%)")
```
